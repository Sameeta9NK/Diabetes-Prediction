```python
##Market Segmentation##
##Segmenting customers into clusters based on their annual income and spending score for developing customized marketing strategies for better customer service and higher revenue
```

```python
## Importing Librariesimport numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
## Acquiring csv data into pandas dataframe
data = pd.read_csv('/content/sample_data/Customers_Mall.csv')
data.head()
```

```python
## Getting information on data
data.info()
```

```python
# Extracting Annual Income and Spending Score column for clustering
X_features = data.iloc[:,[3,4]].values
print(X_features)
```

```python
## Determining number of clusters using elbow method
## Finding W-CSS value for variation of number of clusters

wcss = []

for i in range(1,15):
  kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state= 42)
  kmeans.fit(X_features)

  wcss.append(kmeans.inertia_)
```

```python
## Elbow Graph

sns.set()
plt.plot(range(1,15),wcss)
plt.title('Elbow Graph')
plt.xlabel('#Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## As seen above the optimum number of clusters is 5
## Training k-means clustering model

kmeans = KMeans(n_clusters = 5, init='k-means++', random_state=0)

# Assigning labels for each data point based on their cluster
Y = kmeans.fit_predict(X_features)
print(Y)
```

```python
## Plotting clusters

plt.figure(figsize=(8,8))
plt.scatter(X_features[Y==0,0], X_features[Y==0,1], s=50, c='Red', label='Cluster 1')
plt.scatter(X_features[Y==1,0], X_features[Y==1,1], s=50, c='Pink', label='Cluster 2')
plt.scatter(X_features[Y==2,0], X_features[Y==2,1], s=50, c='Yellow', label='Cluster 3')
plt.scatter(X_features[Y==3,0], X_features[Y==3,1], s=50, c='Green', label='Cluster 4')
plt.scatter(X_features[Y==4,0], X_features[Y==4 ,1], s=50, c='Orange', label='Cluster 5')

#plotting centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='Blue', label='Centroids')

plt.title('Customer Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
